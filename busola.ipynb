3.Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random numbers
random_numbers = np.random.randint(0, 100, 5) 

# Create a sequence of indices for the data points
indices = np.arange(len(random_numbers))

# Create a figure and axis objects
fig, (scatter_ax, plot_ax) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

# Scatter plot
scatter_ax.scatter(indices, random_numbers, color='green', marker='*', label='Data Points', s=100)  # Change color to green, marker to '*', and size to 100
scatter_ax.set_title('Scatter Plot')
scatter_ax.set_xlabel('Index')
scatter_ax.set_ylabel('Value')
scatter_ax.legend()

# line Plot
plot_ax.plot(indices, random_numbers, color='brown', marker='s', linestyle='--', label='Data Points')  # Change color to brown, marker to 's' (square), and linestyle to '--'
plot_ax.set_title('Line Plot')
plot_ax.set_xlabel('Index')
plot_ax.set_ylabel('Value')
plot_ax.legend()

plt.tight_layout()
plt.show()
```
